## 이상치를 찾아라
### 데이터에서 IQR을 활용해 Fare컬럼의 이상치를 찾고, 이상치 데이터의 여성 수를 구하시오
- 강의 영상 : https://youtu.be/ipBW5D_UJEo
- 데이터셋 : titanic 

### IQR이란, InterQuartile Range의 약자로써 Q3 - Q1를 의미한다.

### Q3 - Q1: 사분위수의 상위 75% 지점의 값과 하위 25% 지점의 값 차이



## Hint

# IQR 구하기

### IQR = Q3 - Q1
#### Q1 - 1.5 * IQR
#### Q3 + 1.5 * IQR

# pandas 활용
```
Q1 = df[col].quantile(.25)
Q3 = df[col].quantile(.75)
```
# numpy 활용
```
Q1 = np.percentile(df[col], 25)
Q3 = np.percentile(df[col], 75)
```

## 해답

````python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np
df = pd.read_csv('../빅데이터_분석기사/archive/train.csv')
````

````python
# 간단한 탐색적 데이터 분석 (EDA)
print(df.shape)
print(df.isnull().sum())
df.head()
````

````python
# IQR 구하기
Q1 = df['Fare'].quantile(.25)
Q3 = df['Fare'].quantile(.75)

# numpy 활용시
# Q1 = np.percentile(df['Fare'], 25)
# Q3 = np.percentile(df['Fare'], 75)

IQR = Q3 - Q1
Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
````

### (1)

````python
# 이상치 데이터 구하기
outdata1 = df[df['Fare']<(Q1 - 1.5 * IQR)]
outdata2 = df[df['Fare']>(Q3 + 1.5 * IQR)]
len(outdata1), len(outdata2)
````

````python
# 이상치 데이터에서 여성 수 구하기, 출력하기 print()
print(sum(outdata2['Sex'] == 'female'))
````

### (2)

````python
condition = (df['Fare']<(Q1 - 1.5 * IQR)) | (df['Fare']>(Q3 + 1.5 * IQR))
outdata = df[condition]
len(outdata)
````

````python
len(outdata[outdata['Sex'] == 'female'])
````
